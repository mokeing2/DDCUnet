## 使用数值n=1000000对比python原生之间时差

```python
import time
n=1000000
start_time=time.time()
py_list=list(range(n))
for i in range(n):
    py_list[i]=py_list[i]+1
py_time=time.time()
print(py_time-start_time)
#
a=np.arange(n)
start_time=time.time()
a=a+1
np_time=time.time()
print(np_time-start_time)
```

## 创建一个生成器生成10个整数，用他构建一个数组

```python
def generate():
  for i in range(10):
    yield i#“产生”一个值，然后暂停函数，等待下次被调用时继续
  arr=np.fromiter(generate(),dtype=float) #从一个可迭代对象（比如生成器）中逐个取出元素，组成一个 NumPy 数组
  print(arr)
```

## 如何在向量中找到与给定标量最接近的值

```python
verctor=[1,4,8,2]
target=5
value=verctor[0]
temp=abs(value-target)
for i in verctor:
  res=np.abs(i-target)
  if res<temp:
    temp=res
    value=i
print(value)
```

## 创建一个形状为(10,2)的随机向量，表示坐标，逐点计算距离

```python
arr=np.random.random((10,2))
x,y=np.atleast_2d(arr[:,0],arr[:,1])#把一维数组“升级”成二维数组
d=np.sqrt((x-x.T)**2+(y-y.T)**2) #逆置为了更好使用广播机制
print(d)
```

## 如何按第n列对于数组排序

```python
arr=np.random.randint(0,10,(3,3))
print(arr)
n=1
sorted_arr = arr[arr[:, n].argsort()]
print(sorted_arr)
#arr[:, n]：取出第 n 列的所有数据（得到一维数组）。

#.argsort()：返回这一列数据从小到大排序后的索引位置（比如返回 [2, 0, 1] 表示原数组第2行最小，第0行次之...）。

#arr[...]：按这些索引重新排列原数组的所有行。
```
